> **These commands:**
> * **Request the json results for the "_Civil War_" collection**
> * **The while loop prints the titles from the json results then iterates to the next page if it exists or breaks the loop if it does not**
> * **The "_Civil War_" collection has over 13 thousand entries so the loop takes some time to be completed and the streaming output is limited to the last 5000 lines**

```python
civil_war_json = requests.get("https://www.loc.gov/collections/civil-war/?fo=json").json()
```

```python
while True: #As long as we have a next page, go and fetch it
    for civil_war in civil_war_json["results"]:
        print(civil_war["title"]) #print out the collection's title

    if "pagination" in civil_war_json:
        next_page = civil_war_json["pagination"]["next"] #get the next page url
        if next_page is not None: #make sure we haven't hit the end of the pages
            civil_war_json = requests.get(next_page).json()
        else:
            break #we are done and can stop looping
    else:
        break
```
