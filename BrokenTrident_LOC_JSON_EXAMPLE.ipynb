# Fetching and Printing Titles of Baseball Cards Collections

This Python script fetches the baseball cards collection from the Library of Congress and prints each collection's title.

## Code Explanation

The script starts by sending a GET request to the Library of Congress API to fetch the baseball cards collection:

```python
collections_json = requests.get("https://www.loc.gov/collections/baseball-cards/?fo=json").json()

for collection in collections_json["results"]:
    print(collection["title"])
```

The response from the API is in JSON format. This is converted to a Python dictionary using the .json() method and stored in the collections_json variable.

The script then iterates over each collection in the results list of the collections_json dictionary. For each collection, it prints the title. This allows you to quickly get an overview of the titles of the baseball cards collections from the Library of Congress.

This script is a simple and effective way to programmatically access and display information from the Library of Congress's baseball cards collection.


# Displaying Baseball Cards from the Library of Congress

This Python script fetches the baseball cards collection from the Library of Congress and displays each card's title and front cover art.

The Image and display functions from the IPython.display module are imported. Image is used to create an image object from an image file or URL. Display is used to display the image

```python
from IPython.display import Image, display #tell python that we want to reuse some code to display the images for us

baseball_cards = requests.get("https://www.loc.gov/collections/baseball-cards/?fo=json").json() #get the JSON data

for item in baseball_cards["results"]:
    title = item["title"]
    team = title.split(",")[1].strip()  # Split the title and get the second item
    print(team)
    display(Image(url=item["image_url"][0])) #display the front cover art
```

# Fetching Baseball Cards Collection and Visualizing Team Frequencies

This Python script fetches the baseball cards collection from the Library of Congress, extracts the team names, and creates a bar graph of the team names and their frequencies.

```python
import matplotlib.pyplot as plt
from collections import Counter

# Fetch the baseball cards collection
baseball_cards = requests.get("https://www.loc.gov/collections/baseball-cards/?fo=json").json()

# Initialize an empty list to store the team names
teams = []

# Loop through each item in the results
for item in baseball_cards["results"]:
    # Split the title and get the second item (the team name)
    title = item["title"]
    team = title.split(",")[1].strip()

    # Add the team name to the list
    teams.append(team)

# Count the frequency of each team name
counter = Counter(teams)

# Get the team names and their frequencies
names = list(counter.keys())
frequencies = list(counter.values())

# Create a bar graph
plt.figure(figsize=(20, 10))
plt.bar(names, frequencies)
plt.xlabel('Teams')
plt.ylabel('Frequency')
plt.title('Frequency of Team Names')
plt.show()
```
